### 5-2-1-3 : Multiple plots with subplots()

plt.subplots(행의 갯수, 열의 갯수)

```python
data = np.arange(1, 51)
# data 생성

# 밑 그림
fig, axes = plt.subplots(2, 4)

axes[0, 0].plot(data)
axes[0, 1].plot(data ** 2)
axes[0, 2].plot(data ** 3)
axes[0, 3].plot(data % 10)
axes[1, 0].plot(-data)
axes[1, 1].plot(data // 20)
axes[1, 2].plot(np.sin(data))
axes[1, 3].plot(np.cos(data))

plt.tight_layout()
plt.show()
```

### x, y, colors, area 설정

* colors 는 임의 값을 color 값으로 변환합니다.
* area는 점의 넓이를 나타냅니다. 값이 커지면 당연히 넓이도 커집니다.

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.arange(50)
area = x * y * 2000
```

```python
plt.scatter(x, y, s=area, c=colors)
plt.show()
```

### cmap, alpha 설정

* cmap에 컬러를 지정하면, 컬러 값을 모두 같게 가져갈 수도 있습니다.
* alpha값은 투명도를 나타내며 0 ~ 1 사이의 값을 지정해 줄 수 있으며, 0에 가까울 수록 투명한 값을 가집니다.

```python
np.random.rand(50)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(131)
plt.scatter(x, y, s=area, cmap='yellow', alpha=0.1)
plt.title('alpha=0.1')

plt.subplot(132)
plt.title('alpha=0.5')
plt.scatter(x, y, s=area, cmap='yellow', alpha=0.5)

plt.subplot(133)
plt.title('alpha=1.0')
plt.scatter(x, y, s=area, cmap='yellow', alpha=1.0)

plt.show()
```

## 5-3-2 : Barplot, Barhplot

```python
x = ['수학', '영어', '국어', '과학', '미술', '음악']
y = [91, 70, 56, 67, 75, 45]

plt.figure(figsize=(6, 3))
# plt.bar(x, y)
plt.bar(x, y, align='center', alpha=0.5, color='blue')
plt.xticks(x)
plt.ylabel('점수')
plt.title('과목')

plt.show()
```

### Batplot에서 Multiple Bar 

```python
x_label = ['수학', '영어', '국어', '과학', '미술', '음악']
x = np.arange(len(x_label))
y_1 = [91, 70, 56, 67, 75, 45]
y_2 = [66, 80, 30, 70, 60, 15]
y_3 = [50, 78, 67, 98, 79, 90]

# 폭 지정
width = 0.2

# subplots 생성
fig, axes = plt.subplots()

# Bar 설정
axes.bar(x - width, y_1, width, align='center', alpha=0.5)
axes.bar(x        , y_2, width, align='center', alpha=0.5)
axes.bar(x + width, y_3, width, align='center', alpha=0.5)

# xtick 설정
plt.xticks(x)
axes.set_xticklabels(x_label)
plt.ylabel('점수')
plt.title('과목')

plt.legend(['학생1', '학생2', '학생3'])

plt.show()
```

```python
x_label = ['수학', '영어', '국어', '과학', '미술', '음악']
x = np.arange(len(x_label))
y_1 = [91, 70, 56, 67, 75, 45]
y_2 = [66, 80, 30, 70, 60, 15]
y_3 = [50, 78, 67, 98, 79, 90]

# 넓이 지정
width = 0.2

# subplots 생성
fig, axes = plt.subplots()

# 넓이 설정
axes.barh(x - width, y_1, width, align='center', alpha=0.5, color='red')
axes.barh(x        , y_2, width, align='center', alpha=0.5, color='green')
axes.barh(x + width, y_3, width, align='center', alpha=0.5, color='blue')

# xtick 설정
plt.yticks(x)
axes.set_yticklabels(x_label)
plt.xlabel('점수')
plt.title('과목')

plt.legend(['학생1', '학생2', '학생3'])

plt.show()
```

## 5-3-3 : Line Plot

### Lineplot 

```python
x = np.arange(0, 10, 0.1)
y = 1 + np.sin(x)

plt.plot(x, y)

plt.xlabel('x 값', fontsize=15)
plt.ylabel('y 값', fontsize=15)
plt.title('sin 그래프', fontsize=18)

plt.grid()

plt.show()
```

### Lineplot with multiple lines

* color: 컬러 옵션
* alpha: 투명도 옵션

```python
x = np.arange(0, 10, 0.1)
y_1 = 1 + np.sin(x)
y_2 = 1 + np.cos(x)

plt.plot(x, y_1, label='1+sin', color='red', alpha=0.5)
plt.plot(x, y_2, label='1+cos', color='green', alpha=0.5)

plt.xlabel('x 값', fontsize=15)
plt.ylabel('y 값', fontsize=15)
plt.title('sin cos 그래프', fontsize=18)

plt.grid()
plt.legend()

plt.show()
```

### Lineplot with marker

* marker : 마커 옵션

```python
x = np.arange(0, 10, 0.1)
y_1 = 1 + np.sin(x)
y_2 = 1 + np.cos(x)

plt.plot(x, y_1, label='1+sin', color='red', alpha=0.5, marker='o')
plt.plot(x, y_2, label='1+cos', color='green', alpha=0.5, marker='+')

plt.xlabel('x 값', fontsize=15)
plt.ylabel('y 값', fontsize=15)
plt.title('sin cos 그래프', fontsize=18)

plt.grid()
plt.legend()

plt.show()
```

### Lineplot with linestyle

* linestyle: 라인 스타일 변경 옵션

```python
x = np.arange(0, 10, 0.1)
y_1 = 1 + np.sin(x)
y_2 = 1 + np.cos(x)

plt.plot(x, y_1, label='1+sin', color='red', linestyle=':')
plt.plot(x, y_2, label='1+cos', color='green', linestyle='-.')

plt.xlabel('x 값', fontsize=15)
plt.ylabel('y 값', fontsize=15)
plt.title('sin cos 그래프', fontsize=18)

plt.grid()
plt.legend()

plt.show()
```

## 5-3-6 : Pie Chart 

**Pie Chart 입력 옵션**

* **explode** : Pie에서 빠져 나온 부분의 비율
* **autopct** : 자동으로 퍼센트 표기
* **shadow** : 그림자 표시
* **startangle** : Pie를 그리기 시작하는 각도

**Pie Chart 리턴값**

* **texts** : label에 대한 텍스트 효과를 줄 때 사용한다,

* **autotexts** : Pie 위에 그려지는 텍스트 효과를 줄 때 사용한다.

```python
labels = ['루시드', '웨이모', '테슬라', '애플']
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. '웨이모')

# texts, autotexts 인자를 활용하여 텍스트 스타일링을 적용합니다
patches, texts, autotexts = plt.pie(sizes, 
                                    explode=explode, 
                                    labels=labels,  
                                    autopct='%1.1f%%',
                                    shadow=True, 
                                    startangle=90)

plt.title('전기차 시장 점유율', fontsize=30)

# label 텍스트에 대한 스타일 적용
for t in texts:
    t.set_fontsize(20)
    t.set_color('gray')
    
# pie 위의 텍스트에 대한 스타일 적용
for t in autotexts:
    t.set_color("white")
    t.set_fontsize(18)

plt.show()
```

## 5-3-9 : imshow

이미지 데이터와 같은 2차원의 데이터를 시각화할 때 사용한는 imshow 에 대해서 배운다.

```python
from sklearn.datasets import load_digits

digits = load_digits()
X = digits.images[:10]
X[0]
```

* **load_digits** 는 0~16 값을 가지는 array로 이루어져 있다.

* 1개의 array는 [8 X 8] 배열 안에 표현되어 있다.

* 숫자는 0~9까지 이다.

```python
fig, axes = plt.subplots(nrows=2, ncols=5, sharex=True, figsize=(10, 5), sharey=True)

for i in range(10):
    axes[i//5][i%5].imshow(X[i], cmap='Reds')
    axes[i//5][i%5].set_title('숫자 '+str(i), fontsize=20)
    
plt.tight_layout()
plt.show()
```
