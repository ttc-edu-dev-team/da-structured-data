## .groupby(<font color='green'>key</font>)[<font color='blue'>column</font>].<font color='purple'>operation</font>()


### 같은 <font color='green'>key</font>를 가진 데이터끼리 묶어서
### 그룹마다 <font color='blue'>column</font>을 인덱싱해서(뽑아내서)
### <font color='purple'>operation</font>을 적용하여 계산한 뒤
### 각 그룹의 결과를 합쳐서 하나의 DataFrame으로 만든다.

### groupby 자세히 보기

#### 1st : age열을 gender로 그룹핑

```python
grouped = users['age'].groupby(users.gender)
grouped
```

#### 2nd : Split된 상태로 GroupBy object가 리턴되는데...안을 들여다 볼수는 없다.

#### 3rd : 함수를 적용하면 그룹들의 결과를 합쳐서 보여준다.

```python
grouped.mean()
```

## 1번 방법 : users[<font color='blue'>age</font>].groupby(<font color='green'>users.gender</font>).<font color='purple'>mean</font>()

## 2번 방법 : users.groupby(<font color='green'>gender</font>)[<font color='blue'>age</font>].<font color='purple'>mean</font>()

## 1번 방법이...변수 반복 등 귀찮아서...2번 방법이 나오게 됨.

### groupby 결과를 DataFrame으로 보기

```python
users.groupby('gender')[['age']].mean()
```

## .groupby(key)<font color='red'>[[</font>column<font color='red'>]]</font>.operation()

## groupby와 함께 쓰이는 함수들
- 갯수 .size()
- 합 .sum()
- 평균 .mean()
- 중앙값 .median()
- 표준편차 .std()
- 최대값 .max()
- 최소값 .min()
- 내가 만든 함수 .apply(내가 만든 함수명)

# Lesson 4-5: Method Chaining

## Method Chaining = ?

## Method Chaining = Pandas의 각 기능들을 잘 조합하는 방법

## Pandas 의 수많은 기능들

- 데이터 불러오기
- 정렬하기
- 결측값 제거하기
- 중복값 제거하기
- 컬럼 이름 바꾸기
- DataFrame끼리 사칙연산 하기
- 다른 데이터와 합치기
- ...

## 데이터 분석은 작은 method들의 연속

* a = 데이터에서 원하는 부분 인덱싱
* b = a 와 다른 데이터와 합침
* c = b에서 열 이름 바꿈
* d = c에서 새로운 열 추가함
* e = d를 정렬함
* ...
* z = 최종 내가 원하는 DataFrame

## 문제점
1. 새로운 변수 이름을 생각해 내기 힘들다.
2. 이름만 보고 무슨 변수인지 예측하기 어렵다.
3. 나중에 쓰이지 않을 변수가 너무 많다.

## 해결방법 = Method Chaining

### [Before]
#### a = data.this_method()
#### b = a.that_method()
#### c = b.another_method()

### [After]
#### data_new = <font color='red'>(</font> data.this_method()
####                 .that_method()
####                 .another_method()
####            <font color='red'>)</font>

### 전체를 <font color='red'>소괄호</font>로 묶어야 함.

## Method Chaining 예 : 직업 별로 연령대 비중 구하기

```python
age_to_range = { 1: "Under 18", 18: "18-24", 25: "25-34", 35: "35-44", 45: "45-49", 50: "50-55", 56: "56+" }
to_description = { 0: "other or not specified", 1: "academic/educator", 2: "artist", 3: "clerical/admin",
                4: "college/grad student", 5: "customer service", 6: "doctor/health care",
                7: "executive/managerial", 8: "farmer", 9: "homemaker", 10: "K-12 student", 11: "lawyer",
                12: "programmer", 13: "retired", 14: "sales/marketing", 15: "scientist", 16: "self-employed",
                17: "technician/engineer", 18: "tradesman/craftsman", 19: "unemployed", 20: "writer" }
(users
 .assign(age_range = lambda x: x.age.map(age_to_range))
 .assign(occupation = lambda x: x.occupation.map(to_description))
 .groupby(['occupation', 'age_range']).size()
 .unstack()
 .fillna(0)
 .pipe(lambda x: x.div(x.sum(axis='columns'), axis='index'))
 .mul(100)
 .round(2)
)
```

## 4-6-5 : merge
DataFrame 병합하기

* concat : row나 column 기준으로 단순하게 이어 붙히기
* merge : 특정 고유한 키(unique id) 값을 기준으로 병합하기

```python
df14 = pd.read_csv(filename)
df14.head()
```

```python
df15 = df14.copy()
df15.drop(['Pclass','Name','Sex','Age'], axis=1)
df15.head()
```

```python
df14.drop(['SibSp','Parch','Ticket','Fare','Cabin','Embarked'], axis=1)
df14.head()
```

df14와 df15는 **'PassengerId' 이라는 column이 모두 존재**한다. 

따라서 'PassengerId'을 기준으로 두 DataFrame을 **Merge**할 수 있다.

**pd.merge(left, right, on='Merge의 기준이 되는 column', how='left')**

* left와 right에는 Merge할 두개의 DataFrame을 각각 설정한다.
* on 옵션 : Merge의 기준이 되는 column을 설정한다.
* how 옵션 : 'left', 'right', 'inner', 'outer' 라는 4가지의 Merge 방식중 한가지를 선택한다.

### left, right 옵션 사용 예

```python
df15.head()
```

```python
df_right = df15.drop([1, 2, 3, 4, 5], axis=0)
```

```python
df_right = df_right.reset_index(drop=True)
```

```python
df_right.head()
```

```python
df_left = df14.copy()
```

```python
pd.concat([df_left, df_right], axis=1)
```

#### concat을 사용해서 합쳤을 때는 제대로 합쳐지지 않는다

```python
pd.merge(df_left, df_right, on='PassengerId', how='left')
```

#### 'left' 옵션일 경우, left DataFrame에 키 값이 존재하면 해당 데이터를 유지하고, Merge한 right DataFrame의 값은 NaN이 된다.

```python
pd.merge(df_left, df_right, on='PassengerId', how='right')
```

#### 'right' 옵션일 경우, right DataFrame을 기준으로 Merge하게 된다.

이 예에서는 left DataFrame이 더 많은 데이터를 갖고 있기 때문에, right를 기준으로 Merge한 결과 DataFrame의 행 크기가 줄어든다.

### Inner, Outer 옵션 사용 예
* inner 옵션 : 두 DataFrame 모두 Key가 존재하는 경우만 Merge한다.
* outer 옵션 : 하나의 DataFrame에 Key가 존재하는 경우 모두 Merge한다.
* outer 옵션 선택시, 없는 값은 NaN이 된다.

```python
pd.merge(df_left, df_right, on='PassengerId', how='inner')
```

```python
pd.merge(df_left, df_right, on='PassengerId', how='outer')
```
