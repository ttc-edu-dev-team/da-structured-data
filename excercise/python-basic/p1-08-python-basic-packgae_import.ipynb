## 2. 모듈 import 하기

### (여기서 잠깐) import 는 왜 할까?

**import**는 파이썬의 유용한 기능 중 하나입니다.

**import** 는 다른 사람이 이미 만들어 놓은 좋은 모듈을 우리가 쉽게 가져다 쓸 수 있습니다.

### import 하는 방법

.py (파이썬 파일 확장자)로 된 파일을 우리는 **모듈** 이라고 한다고 했었는데요, **import** 구문을 통해 해당 파일을 불러올 수 있습니다.

```python
import pandas
```

위의 코드는 pandas라는 모듈을 우리가 불러오겠다라는 의미이며, 이미 colab에서 유명한 모듈은 설치가 되어 있는 상태이기 때문에, 우리가 따로 설치할 필요없이 가져올 수 있습니다.

## 3. 패키지 에서 import 하기

패키지는 모듈의 그룹입니다.<br>
폴더 구조를 떠올리시면 전혀 어렵지 않습니다<br>
자, 그럼 폴더 구조로 우리가 파일을 관리한다고 가정해 봅시다.

**[예시]**
파일1, 파일2, 파일3 을 하나의 '직박구리'라는 폴더 안에 넣어놨다고 가정해 봅시다. 
* '직박구리' 폴더는 패키지
* 파일1, 파일2, 파일3는 각각의 모듈이 됩니다.

우리는 **직박구리 > 파일1** 모듈을 가져오기 위해서는 다음과 같이 불러올 수 있습니다.

**from** 직박구리 **import** 파일1
